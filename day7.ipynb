```python
from copy import deepcopy
with open("input.txt") as file:
  lines=[line.rstrip('\n') for line in file]
  print(lines)
```

```python
from collections import defaultdict
dict1=defaultdict(int)
path=[]
for line in lines:
    cmd = line.strip().split()
    if cmd[1] == 'cd':
        if cmd[2] == '..':
            path.pop()
        else:
            path.append(cmd[2])
    elif cmd[1] == 'ls':
        continue
    elif cmd[0] == 'dir':
        continue
    else:
        size = int(cmd[0])
        print(path)
        for i in range(1, len(path)+1):
            dict1['/'.join(path[:i])] += size


```

```python
sum = 0
minimum=9999999999999999999999999
threshold=dict1['/']-40000000
print(threshold)
for key,size in dict1.items():
    if size <= 100000:
        sum += size
    
    if size>=threshold:
        minimum=min(minimum,size)


print(sum)
print(minimum)
```
